```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time
import os
from collections import defaultdict
import matplotlib.pyplot as plt
import pickle
```

```python
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import Select
```

```python
chromedriver = "/Users/kendra/ds/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
driver = webdriver.Chrome(chromedriver)
```

```python
url = 'https://projects.fivethirtyeight.com/2018-nfl-predictions/'
```

```python
elo_dict = defaultdict(dict)
driver.get(url)
time.sleep(3)

for val in range(6,0,-1):
    print(val)
    
    if val != 6:
        select = Select(driver.find_element_by_tag_name('select'))
        select.select_by_value(str(val))
        time.sleep(3)
        
    elo_els = driver.find_elements_by_xpath("//td[@class='num elo']")
    elos = [int(x.text) for x in elo_els]

    table = driver.find_elements_by_xpath("//tr")
    team_els = table[4:(4+32)]
    teams = [x.get_attribute('data-team') for x in team_els]
    
    elo_dict[val] = dict(zip(teams, elos))

driver.close()
```
